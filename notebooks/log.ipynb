```python
import pandas as pd
import matplotlib.pylab as plt
log_file = "../data/ram.log"
df = pd.read_csv(log_file, sep=",")
# drop some points as the file is too large
df = df.iloc[::2, :]
df["MB"] = df["RSS_KB"] / 1024

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(df["Timestamp"], df["MB"], label="RAM Usage (MB)")
ax.set_xlabel("Time (s)")
ax.set_ylabel("RAM Usage (MB)")
ax.set_title("RAM Usage Over Time")
plt.show()
```
