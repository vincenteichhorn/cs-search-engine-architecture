```python
from matplotlib import pyplot as plt
import numpy as np
import torch
import torch.nn.functional as F

ranks = np.array(list(range(1, 11)))
relevances = torch.tensor(ranks.max() - ranks + 1).unsqueeze(0).float()
temp = 1.5
target_probs = F.softmax(relevances / temp, dim=1)

#bar plot
plt.bar(ranks, target_probs.numpy().flatten())
plt.xlabel("Rank")
plt.ylabel("Probability")
plt.title("Target Probability Distribution over Ranks")
plt.show()
```
