```python
from matplotlib import pyplot as plt
import numpy as np
import torch
import torch.nn.functional as F

ranks = np.array(list(range(1, 50)))
relevances = torch.tensor(ranks.max() - ranks + 1).unsqueeze(0).float()
temp = 0.7
target_probs = relevances# F.softmax(relevances / temp, dim=1)
target_probs = F.softmax(relevances / temp, dim=1)

#bar plot
plt.bar(ranks, target_probs.numpy().flatten())
plt.xlabel("Rank")
plt.ylabel("Relevance Score")
plt.title("Target Relevance Score Distribution over Ranks")
plt.show()

# random
model_probs = torch.rand_like(target_probs) 
model_probs = F.softmax(model_probs, dim=1)

#bar plot
plt.bar(ranks, model_probs.numpy().flatten())
plt.xlabel("Rank")
plt.ylabel("Relevance Score")
plt.title("Target Relevance Score Distribution over Ranks")
plt.show()
```
